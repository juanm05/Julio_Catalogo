```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("Julio_Catalogo_Region.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
# Regiones con mas link clicks 
temp_df = df.groupby("Region")[["Link clicks", "Impressions", "Reach"]].sum().reset_index().sort_values(by = "Link clicks", ascending = False).reset_index(drop = True)

temp_df["Result rate"] = temp_df["Link clicks"] * 100 / temp_df["Reach"]

temp_df.sort_values(by = "Result rate", ascending = False)

freq_df = df.groupby("Region")["Frequency"].mean().reset_index()

temp_df = pd.merge(temp_df, freq_df, on = "Region", how = "left")

temp_df[:20]
```

```python
# Cost Per Link Click - Amount Spent 
df_amount = df.groupby("Region")["Amount spent (MXN)"].sum().reset_index()

df_cpc = df.groupby("Region")["CPC (cost per link click) (MXN)"].mean().reset_index()

df_union = pd.merge(df_amount, df_cpc, how = "left", on = "Region")

temp_df = pd.merge(df_union.reset_index(drop = True), temp_df, on = "Region")

temp_df.sort_values(by = "CPC (cost per link click) (MXN)")[:20].reset_index(drop = True)
```
